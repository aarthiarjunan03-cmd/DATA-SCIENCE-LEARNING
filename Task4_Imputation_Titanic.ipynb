# Imputation - Titanic Dataset

```python
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('4_Imputation_Titanic.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
age_fill = SimpleImputer(strategy='median')
df['age'] = age_fill.fit_transform(df[['age']])
print('filled age with:', age_fill.statistics_)
```

```python
embarked_fill = SimpleImputer(strategy='most_frequent')
df['embarked'] = embarked_fill.fit_transform(df[['embarked']]).ravel()
print('filled embarked with:', embarked_fill.statistics_)
```
