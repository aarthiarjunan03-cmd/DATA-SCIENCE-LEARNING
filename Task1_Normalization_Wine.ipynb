# Normalization - Wine Dataset

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('1_Normalization_Wine.csv')
df.head()
```

```python

df[['malic_acid','proline','magnesium']].describe()
```

```python
cols = ['malic_acid','proline','magnesium']

scaler = MinMaxScaler()
df[cols] = scaler.fit_transform(df[cols])

df.head()
```

```python
df[cols].describe()
```

```python
fig, ax = plt.subplots(1,3, figsize=(14,4))
for i,c in enumerate(cols):
    ax[i].hist(df[c], bins=15, color='skyblue', edgecolor='black')
    ax[i].set_title(c)
plt.tight_layout()
plt.show()
```
