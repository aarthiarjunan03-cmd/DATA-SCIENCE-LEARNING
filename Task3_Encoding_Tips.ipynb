# Encoding - Tips Dataset

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('3_Encoding_Tips.csv')
df.head()
```

```python
print(df['day'].unique())
print(df['sex'].unique())
print(df['time'].unique())
```

```python
# one hot encoding for nominal columns
df_encoded = pd.get_dummies(df, columns=['day','sex'])
df_encoded.head()
```

```python
# manual mapping for time since order matters (lunch -> dinner)
time_order = {'Lunch':0, 'Dinner':1}
df_encoded['time'] = df['time'].map(time_order)
df_encoded.head()
```

```python
le = LabelEncoder()
check = le.fit_transform(df['time'])
print(list(le.classes_))
print(check[:10])
```
